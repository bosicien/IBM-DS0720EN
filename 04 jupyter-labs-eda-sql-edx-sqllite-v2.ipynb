### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# Task 1: Display the names of the unique launch sites
query = 'SELECT DISTINCT "Launch_Site" FROM SPACEXTBL'

# Executing using the cursor or pandas
cur.execute(query)
launch_sites = cur.fetchall()

for site in launch_sites:
    print(site[0])
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
# SQL query to select records starting with 'KSC'
query = 'SELECT * FROM SPACEXTBL WHERE "Launch_Site" LIKE "KSC%" LIMIT 5'

# Execute and display the results using Pandas for a clean table format
import pandas as pd
df = pd.read_sql_query(query, con)
print(df)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Task 3: Display the total payload mass carried by boosters launched by NASA (CRS)
query = 'SELECT SUM("PAYLOAD_MASS__KG_") FROM SPACEXTBL WHERE "Customer" = "NASA (CRS)"'

cur.execute(query)
total_payload = cur.fetchone()[0]

print(f"Total Payload Mass for NASA (CRS): {total_payload} kg")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Task 4: Display average payload mass carried by booster version F9 v1.1
query = 'SELECT AVG("PAYLOAD_MASS__KG_") FROM SPACEXTBL WHERE "Booster_Version" = "F9 v1.1"'

cur.execute(query)
avg_payload = cur.fetchone()[0]

print(f"Average Payload Mass for F9 v1.1: {avg_payload:.2f} kg")
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
# Task: List the earliest date where a successful landing outcome in drone ship was achieved
query = 'SELECT MIN("Date") FROM SPACEXTBL WHERE "Landing_Outcome" = "Success (drone ship)"'

cur.execute(query)
# Accessing the first element of the returned tuple
first_drone_success_date = cur.fetchone()[0]

print(f"Date of first successful drone ship landing: {first_drone_success_date}")
```

### Task 6
### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
query = """
SELECT Booster_Version 
FROM SPACEXTBL 
WHERE ("Landing _Outcome" LIKE '%Success%ground%pad%' 
   OR "Landing_Outcome" LIKE '%Success%ground%pad%')
  AND PAYLOAD_MASS__KG_ > 4000 
  AND PAYLOAD_MASS__KG_ < 6000
"""

cur.execute(query)
boosters = cur.fetchall()

if not boosters:
    print("Still no records. Check if the Payload Mass is exactly 4000 or 6000.")
else:
    for booster in boosters:
        print(f"Booster Version: {booster[0]}")
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# Task: List the total number of successful and failure mission outcomes
query = 'SELECT "Mission_Outcome", COUNT(*) FROM SPACEXTBL GROUP BY "Mission_Outcome"'

cur.execute(query)
outcomes = cur.fetchall()

for outcome in outcomes:
    print(f"Outcome: {outcome[0]} | Count: {outcome[1]}")
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python

# Task: List all booster_versions that have carried the maximum payload mass
query = 'SELECT "Booster_Version" FROM SPACEXTBL WHERE "PAYLOAD_MASS__KG_" = (SELECT MAX("PAYLOAD_MASS__KG_") FROM SPACEXTBL)'

cur.execute(query)
max_payload_boosters = cur.fetchall()

print("Booster Versions with Maximum Payload Mass:")
for booster in max_payload_boosters:
    print(booster[0])
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
# Task: List records with month names, ground pad successes, and booster versions for 2017
# Using LIKE to avoid exact string match issues
query = """
SELECT 
    CASE 
        WHEN Date LIKE '%-01-%' THEN 'January'
        WHEN Date LIKE '%-02-%' THEN 'February'
        WHEN Date LIKE '%-03-%' THEN 'March'
        WHEN Date LIKE '%-04-%' THEN 'April'
        WHEN Date LIKE '%-05-%' THEN 'May'
        WHEN Date LIKE '%-06-%' THEN 'June'
        WHEN Date LIKE '%-07-%' THEN 'July'
        WHEN Date LIKE '%-08-%' THEN 'August'
        WHEN Date LIKE '%-09-%' THEN 'September'
        WHEN Date LIKE '%-10-%' THEN 'October'
        WHEN Date LIKE '%-11-%' THEN 'November'
        WHEN Date LIKE '%-12-%' THEN 'December'
    END AS Month_Name,
    "Landing_Outcome", 
    "Booster_Version", 
    "Launch_Site"
FROM SPACEXTBL
WHERE "Landing_Outcome" LIKE '%Success (ground pad)%' 
  AND Date LIKE '%2017%'
"""

cur.execute(query)
records_2017 = cur.fetchall()

if not records_2017:
    print("Still no records found. Check if 'Landing_Outcome' contains the word 'Success'.")
else:
    for row in records_2017:
        print(row)
""
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python

# Task: Rank landing outcomes between 2010-06-04 and 2017-03-20
query = """
SELECT "Landing_Outcome", COUNT(*) AS Outcome_Count
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY "Landing_Outcome"
ORDER BY Outcome_Count DESC
"""

cur.execute(query)
ranked_outcomes = cur.fetchall()

print("Ranked Landing Outcomes (2010-06-04 to 2017-03-20):")
for outcome in ranked_outcomes:
    print(f"{outcome[0]}: {outcome[1]}")
```
